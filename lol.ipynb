# API WORK

## Connecting to API (related job)

```python
response = requests.get('http://ddragon.leagueoflegends.com/cdn/11.5.1/data/es_ES/champion.json')
```

## Getting json

```python
champ_ls = list(response.json()['data'].keys())
```

```python
champ_df = pd.DataFrame()
for i in range(len(champ_ls)):
    champ_data = pd.json_normalize(response.json()['data'][champ_ls[i]])
    champ_df = champ_df.append(champ_data)
```

```python
champ_df.to_csv('champions.csv')
```

```python
item_req = requests.get('http://ddragon.leagueoflegends.com/cdn/11.5.1/data/es_ES/item.json')

item_ls = []
for i in list(range(0,10000)):
    try:
        a = item_req.json()['data'][str(i)]
        item_ls.append(str(i))
    except:
        pass
```

```python
item_table = pd.DataFrame()
for i in item_ls:
    item_id = i
    try:
        name = item_req.json()['data'][i]['name']
    except:
        name = np.nan
        
    try:
        upper_item = item_req.json()['data'][i]['into']
    except:
        upper_item = np.nan
    
    try:
        explain = item_req.json()['data'][i]['plaintext']
    except:
        explain = np.nan
    
    try:
        buy_price = item_req.json()['data'][i]['gold']['base']
    except:
        buy_price = np.nan
    
    try:
        sell_price = item_req.json()['data'][i]['gold']['sell']
    except:
        sell_price = np.nan
        
    try:
        tag = item_req.json()['data'][i]['tags'][0]
    except:
        tag = np.nan
    
    pre_df = pd.DataFrame({'item_id' : [item_id],
                           'name' : [name],
                           'upper_item' : [upper_item],
                           'explain' : [explain],
                           'buy_price' : [buy_price],
                           'sell_price' : [sell_price],
                           'tag' : [tag]
                          })
    
    item_table = item_table.append(pre_df)
```
